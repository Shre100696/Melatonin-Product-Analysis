```python
import re
import pandas as pd

# create an example DataFrame
df = pd.DataFrame(product_title_asin['product_title'].unique(), columns=['product_title'])

# define a regular expression to extract mg values
mg_regex = re.compile(r'\d+\s?mg|\d+mg')

# define a function to extract mg values from a product name
def extract_mg_value(text):
    mg_match = mg_regex.search(text)
    if mg_match:
        mg_value = mg_match.group().strip()
        return mg_value
    else:
        return None

# apply the function to the product_name column and append the results as a new column
df['mg_value'] = df['product_title'].apply(lambda x: extract_mg_value(x))

# print the updated DataFrame
print(df)
```

```python
merged_df = pd.merge(df, df2, on='product_title')
pd.set_option('display.max_columns',4)
merged_df

merged_df['mg_value'] = df['mg_value'].str.extract(r'(\d+)').fillna(0).astype('int')

merged_df
```

```python
filtered_df = merged_df[merged_df['Type']=='Adults']
filtered_df
kids_filtered_df = merged_df[merged_df['Type']=='Kids']
kids_filtered_df
plt.barh(filtered_df['product_title'],filtered_df['mg_value'], color='green')
plt.barh(kids_filtered_df['product_title'],kids_filtered_df['mg_value'], color='red')
```
